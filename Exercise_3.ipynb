# üß† Exercise 3: Visualizing Activation Functions


**Objective:**  
Understand how different activation functions (ReLU, Sigmoid, Tanh) transform data and affect learning in neural networks.


**üìò Explanation of Each Activation:**  

| Activation | Formula | Characteristics |
|---|---|---|
| **ReLU**   | `f(x) = max(0, x)` | Fast, sparse output, widely used in CNNs |
| **Sigmoid**| `f(x) = 1 / (1 + e^-x)` | Smooth curve, used in binary classification, suffers from vanishing gradients |
| **Tanh**   | `f(x) = (e^x - e^-x)/(e^x + e^-x)` | Output in [-1, 1], zero-centered, still can vanish |


### **üì¶ Step 1: Import Libraries**


```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

### **üìà Step 2: Define Input Range**

```python
x = np.linspace(-10, 10, 100) # [-10, ..., 10] - total 1000 values
```

### **‚öôÔ∏è Step 3: Define Activation Functions**

```python
# Using NumPy for visual comparison
relu = np.maximum(0, x)
sigmoid = 1 / (1 + np.exp(-x))
tanh = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

### **üé® Step 4: Plot the Activation Functions**

```python
plt.figure(figsize=(12, 4))

# ReLU
plt.subplot(1, 3, 1)
plt.plot(x, relu, label="ReLU", color="red")
plt.title("ReLU")
plt.grid(True)

# Sigmoid
plt.subplot(1, 3, 2)
plt.plot(x, sigmoid, label="Sigmoid", color="green")
plt.title("Sigmoid")
plt.grid(True)

# Tanh
plt.subplot(1, 3, 3)
plt.plot(x, tanh, label="Tanh", color="blue")
plt.title("Tanh")
plt.grid(True)

plt.tight_layout()
plt.show()
```
